```python
import numpy as np
import matplotlib.pyplot as plt
import math

lam = 3009/100  # параметр λ

# Максимальное значение k для которого будем рассчитывать вероятность
max_k = 60
# Создаем список k значений от 0 до max_k
k_values = np.arange(0, max_k+1)

# Расчет вероятностей для каждого k с помощью формулы Пуассона
probs = [(math.exp(-lam) * lam**k) / math.factorial(k) for k in k_values]

# Расчет суммарной вероятности для каждого k
cum_probs = np.cumsum(probs)

# Выбор цвета каждого столбца на основе порога вероятности в 95%
colors = ['red' if x < 0.95 else 'blue' for x in cum_probs]

# Построение графика
plt.bar(k_values, probs, color=colors)
plt.title('Распределение Пуассона')
plt.xlabel('k')
plt.ylabel('P(k)')
plt.show()
```
